## Eliminemos duplicados 

- Borramos los duplicados en los registros correspondientes a 50 registros de Venezuela y 1 de Chile 

```python
cancionesdb2=cancionesdb.drop_duplicates(subset=nombres)
cancionesdb2.info()
```

- Eliminamos los registros donde la variable **cancion_numero** es mayor que 50

```python
grupos_paises=cancionesdb2.groupby('pais_nombre')
print(grupos_paises['cancion_numero'].agg([np.max, np.min, np.sum]))
```

## Sacando información

- ¿Cuál es la canción con más popularidad?

```python
cancionesdb2.sort_values(['cancion_popularidad','cancion_nombre','cancion_artista'], ascending=False)[['cancion_nombre','cancion_artista','cancion_numero','pais_nombre','cancion_popularidad']].head(10)
```

- ¿Cuáles son las tres canciones más populares por pais?

```python
grupos=cancionesdb2.sort_values('cancion_nombre').groupby('pais_nombre')
for name,group in grupos:
   print(name,':\n')
   print(group[['cancion_nombre','cancion_artista']].head(3))
   print("\n")
    
```

- ¿Hay algún artista que tenga más de una canción en los rankings de top 50?

```python
data_artistas=cancionesdb2.drop_duplicates(['cancion_artista','cancion_nombre'])
artistas_masDeUnaCanción=data_artistas.groupby('cancion_artista').size()
artistas_masDeUnaCanción.sort_values(ascending=False)[artistas_masDeUnaCanción>1]
```

- Veamos la distribucion de algunas de las características de las canciones 

```python
canciones=cancionesdb2.sort_values(['cancion_popularidad','cancion_nombre','cancion_artista'], ascending=False).drop_duplicates(['cancion_nombre','cancion_artista'])
canciones.head(5)
```

```python
g = sns.distplot(canciones.feature_bailabilidad,label='bailabilidad')
sns.distplot(canciones.feature_energia,label='energia')
sns.distplot(canciones.feature_positividad,label='positividad')
sns.distplot(canciones.feature_hablado,label='hablado')
g.set_xlim(0, 1)
plt.legend()
```

```python
plt.xticks(rotation = 90)
paises = cancionesdb2.groupby('pais_nombre').mean().sort_values('feature_bailabilidad', ascending=False).index
sns.barplot(x="pais_nombre", y="feature_bailabilidad", order=paises, data=cancionesdb2);
```

```python
plt.xticks(rotation = 90)
paises = cancionesdb2.groupby('pais_nombre').mean().sort_values('feature_hablado', ascending=False).index
sns.barplot(x="pais_nombre", y="feature_hablado", order=paises, data=cancionesdb2);
```

```python
plt.xticks(rotation = 90)
paises = cancionesdb2.groupby('pais_nombre').mean().sort_values('feature_fuerza', ascending=False).index
sns.boxplot(x="pais_nombre", y="feature_fuerza", order=paises, data=cancionesdb2);
```
